```python
losses = {'loss':[],'val_loss':[]}
with open('output_train.log') as log:
    text_file = log
    for line in text_file.readlines():
        if 'val_loss' in line:
            split = line.split()
            losses['loss'].append(float(split[16]))
            losses['val_loss'].append(float(split[18]))
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
ax.plot(losses['val_loss'], label = 'Validation Loss')
ax.plot(losses['loss'], label = 'Training Loss')
ax.set_xlabel('Iterations')
ax.set_ylabel('Loss Value')
ax.set_title('Loss Values During Training')
ax.legend()
```
